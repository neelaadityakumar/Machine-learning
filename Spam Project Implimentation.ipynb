```python
import numpy as np
import pandas as pd
```

```python
#Converting csv to DF
df = pd.read_csv('spam.csv', encoding='latin-1')
```

```python
df
```

```python
#Getting stopword

stopwords = []

def get_the_sw():
    with open('stopwords.txt') as stops:
        for i in stops:
            stopwords.append(i.strip())
get_the_sw()
stopwords
```

```python
df.v2
```

```python
len(df.v2)
```

```python
mails = []

def get_the_corpus():
    for i in df.v2:
        mail = i.strip().split()
        mail = [word for word in mail if word not in stopwords]
        mails.append(mail)
get_the_corpus()
mails
```

```python
from collections import defaultdict
frequency = defaultdict(int)
def removal_singal_freq():
    for mail in mails:
        for word in mail:
            frequency[word] += 1
removal_singal_freq()
mails = [[word for word in mail if frequency[word]>1]for mail in mails]
mails
```

```python
import gensim
dictionary = gensim.corpora.Dictionary(mails)
b_of_w = [ dictionary.doc2bow(mail) for mail in mails]
n_un = len(dictionary)
dense_bow = gensim.matutils.corpus2dense(b_of_w, num_terms = n_un).transpose()
datset = dense_bow
```

```python
len(datset)
```

```python
target = []
for i in df.v1:
    if i == 'ham':
        target.append(1)
    elif i == 'spam':
        target.append(0) 
target
```

```python
import pickle
file_vec = open('datset_vec.pkl','rb')
dataset = pickle.load(file_vec)
file_vec.close()

dataset
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10,shuffle=True)
```

```python
from sklearn import svm
from sklearn import naive_bayes as nb
from sklearn.metrics import f1_score
from numpy import mean
accuracies = []
scores = []
```

```python
for train, test in kf.split(dataset):
    print("1")
    train_set = []
    train_labels = []
    test_set = []
    test_labels = []
    for i in train:
        train_set.append(dataset[i])
        train_labels.append(target[i])
    for i in test:
        test_set.append(dataset[i])
        test_labels.append(target[i])
    classifier = svm.SVC()
#     classifier = nb.GaussianNB()
 #   classifier = nb.MultinomialNB()
#     classifier = nb.BernoulliNB()
    predicted = classifier.fit(train_set, train_labels).predict(test_set)
    score = f1_score(test_labels, predicted, average = 'weighted')
    scores.append(score)
    incorrect = (test_labels != predicted).sum()
    accuracy = (len(test_set) - incorrect) / len(test_set) * 100.
    accuracies.append(accuracy)
#     print("1")
    print("Maximum accuracy attained ",accuracy,accuracies)
#print("f1score  ", scores[np.argmax(accuracies)])
    print('\n')
```
